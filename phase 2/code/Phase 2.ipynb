# Read data

```python
import json

with open('data/train.json') as file:
    train_data = json.load(file)

with open('data/validation.json') as file:
    val_data = json.load(file)
```

# Trie Data Structure

```python
class Node:
    def __init__(self):
        self.mark = False
        self.edges = [None] * 57
        self.appears_docs = []
        
def trie_insert(node, string, doc_num, idx = 0): 
    
    if idx == len(string): 
        node.mark = True
        if doc_num not in node.appears_docs:
            node.appears_docs.append(doc_num)
        return 
    
    while (ord(string[idx])-ord('A')) not in range(57):
        idx += 1
        if len(string) == idx:
            trie_insert(node, string, doc_num, idx)
            return
        
    e = ord(string[idx])-ord('A')
    
    if node.edges[e] is None:
        node.edges[e] = Node()
        
    trie_insert(node.edges[e], string, doc_num, idx+1)

def trie_search(node, string, idx = 0):
    
    if idx == len(string):
        return len(node.appears_docs)
    
    while (ord(string[idx])-ord('A')) not in range(57):
        idx += 1
        if len(string) == idx:
            return trie_search(node, string, idx)
        
    e = ord(string[idx])-ord('A')
    
    if node.edges[e] is None:
        return False
    
    return trie_search(node.edges[e], string, idx+1)

root = Node()
```

# Preprocessing

```python
import numpy as np
import string

N = len(train_data)
docs_train = [None for _ in range(N)]
cat_train = [None for _ in range(N)]
all_words = list()

remove_punc = str.maketrans(string.punctuation, ' '*len(string.punctuation))

for i in range(N):
    cat_train[i] = train_data[i]['category']
    words, tf = np.unique((train_data[i]['title']+" "+train_data[i]['body']).lower().
                          translate(remove_punc).split(), return_counts=True)
    
    docs_train[i] = dict(zip(words, tf))
    all_words += words.tolist()
    
    for word in words:
        trie_insert(root, word, i)
    print("doc num " + str(i) + " processing done!")
```

# calculate X_train

```python
uniqe_words = np.unique(all_words)
N_words = len(uniqe_words)
vocab = dict(zip(uniqe_words, range(N_words)))

X_train = np.zeros((N, N_words))

for i in range(len(docs_train)):
    for word in list(docs_train[i]):
        df = trie_search(root, word)
        X_train[i][vocab[word]] = docs_train[i][word] * np.log10(N / df)
```

# calculate X_val

```python
N_val = len(val_data)
docs_val = [{} for _ in range(N_val)]
cat_val = [None for _ in range(N_val)]

X_val = np.zeros((N_val, N_words))

for i in range(N_val):
    
    cat_val[i] = val_data[i]['category']
    words, tf = np.unique((val_data[i]['title']+" "+val_data[i]['body']).lower()
                          .translate(remove_punc).split(), return_counts=True)
    
    for j in range(len(words)):
        if words[j] in vocab:
            value = trie_search(root, words[j])
            docs_val[i][words[j]] = tf[j]
            X_val[i][vocab[words[j]]] = tf[j] * np.log10(N / value)
```

```python
cat_train, cat_val = np.array(cat_train), np.array(cat_val)
```

# kNN vectorized Implemention

```python
def kNN(X_train, X_val, k = 1, distance_measure = 'Euclidean'):
    
    num_tr = X_train.shape[0]
    num_val = X_val.shape[0]
    dists = np.zeros((num_val, num_tr))
    
    if distance_measure == 'Cosine':
        
        norms = np.linalg.norm(X_train, axis = 1)
        X_norm = (X_train.T / norms).T

        val_norms = np.linalg.norm(X_val, axis = 1)
        X_valNorm = (X_val.T / val_norms).T

        dists = np.matmul(X_valNorm, X_norm.T)
        temp = np.reshape(cat_train[np.argpartition(dists, -k, axis=1)[:, -k:].flatten()], (num_val, k))
        axis = 1
        u, indices = np.unique(temp, return_inverse=True)
        cat_pred = u[np.argmax(np.apply_along_axis(np.bincount, axis, indices.reshape(temp.shape),
                                None, np.max(indices) + 1), axis=axis)]
        
    if distance_measure == 'Euclidean':
        
        N_train = np.tile(np.linalg.norm(X_train, axis=1)**2, (num_val, 1))
        N_val = np.tile(np.linalg.norm(X_val, axis=1)**2, (num_tr, 1)).T
        dists = np.sqrt(np.abs(N_train + N_val - 2 * np.matmul(X_val, X_train.T)))
        
        temp = np.reshape(cat_train[np.argpartition(dists, k, axis=1)[:, :k].flatten()], (num_val, k))
        axis = 1
        u, indices = np.unique(temp, return_inverse=True)
        cat_pred = u[np.argmax(np.apply_along_axis(np.bincount, axis, indices.reshape(temp.shape),
                                None, np.max(indices) + 1), axis=axis)]
        
    return cat_pred
```

# Evaluation

```python
def evaluation(pred, k, measure, alpha, methods = ["acc"]):
    
    if alpha == None:
        print("\n Evaluation with k = " + str(k) + ", distance measure: " + measure + ":\n")
    else:
        print("\n Evaluation with alpha = " + str(alpha) + " :\n")
    
    if "acc" in methods:
        accuracy = np.sum(pred == cat_val) / N_val
        print("   > accuracy", accuracy)
        print()
        
    if "PR" in methods:
        tp = [0 for _ in range(5)]
        fp = [0 for _ in range(5)]
        fn = [0 for _ in range(5)]
        FI = [0 for _ in range(5)]

        for i in range(N_val):
            if pred[i] == cat_val[i]:
                tp[cat_val[i]] += 1
            else:
                fp[int(pred[i])] += 1
                fn[cat_val[i]] += 1

        for i in range(1, 5):
            p = tp[i] / (tp[i] + fp[i])
            r = tp[i] / (tp[i] + fn[i])
            FI[i] = (2 * p * r) / (p + r)

            print("   > Class Num", i)
            print("   Precision:", p)
            print("   Recall:", r)
            print()
    
    if "conf_matrix" in methods:
        confusion_matrix = np.zeros((5, 5))
        for i in range(N_val):
            confusion_matrix[cat_val[i]][int(pred[i])] += 1

        print("   > confusion_matrix:\n", confusion_matrix[1:,1:])
    
    if "macro" in methods:
        macro_F1 = sum(FI) / 4
        print("\n   > macro F1 with beta = 1:", macro_F1)
        
    print("**************")
    
    return
```

```python
for k in [1, 3, 5]:
    for measure in ['Euclidean', 'Cosine']:
        kNN_pred = kNN(X_train, X_val, k = k, distance_measure = measure)
        evaluation(kNN_pred, k, measure, None)
```

```python
best_k, best_measure = 5, "Cosine"
evaluation(kNN_pred, best_k, best_measure, None, methods=["acc", "PR", "conf_matrix", "macro"])
```

# Nave Bayes

```python
def train_NB(alpha, trainDocs, vocab): 
    text_c = [{} for _ in range(5)]
    prior, numWords = np.zeros(5), np.zeros(5)
    N_term = len(vocab)
    condprob = np.zeros((N_term, 5))

    for i in range(len(trainDocs)):
        for word in trainDocs[i]:
            if word in text_c[cat_train[i]]:
                text_c[cat_train[i]][word] += trainDocs[i][word]
            else:
                text_c[cat_train[i]][word] = trainDocs[i][word]

            numWords[cat_train[i]] += trainDocs[i][word]

        prior[cat_train[i]] += 1

    prior /= len(trainDocs)

    for word in list(vocab):
        for class_ in range(1, 5):
            if word in text_c[class_]:
                condprob[vocab[word]][class_] = (text_c[class_][word] + alpha) / (numWords[class_] + alpha * N_term)
            else:
                condprob[vocab[word]][class_] = (alpha / (numWords[class_] + alpha * N_term))
                
    return prior, condprob
```

```python
def apply_NB(valDocs, prior, condprob, vocab):
    pred = np.zeros(N_val)
    for i in range(N_val):
        score = np.array([0] + np.log10(np.array(prior[1:])).tolist())
        for class_ in range(1, 5):
            for term in valDocs[i]:
                if term in vocab:
                    score[class_] += np.log10(condprob[vocab[term]][class_])
                    
        pred[i] = np.argmax(score[1:]) + 1
    
    return pred
```

```python
alphas = np.random.uniform(0, 100, 20)
for alpha in alphas:
    p, prob = train_NB(alpha, docs_train, vocab)
    NB_pred = apply_NB(docs_val, p, prob, vocab)
    evaluation(NB_pred, None, None, alpha, methods=['acc'])
```

# KNN + nltk (stemming)

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

root_ = Node()
tf = [None for _ in range(N)]
vocab_ = list()

for i in range(N):
    words = (train_data[i]['title']+" "+train_data[i]['body']).lower().translate(remove_punc).split()
    for j in range(len(words)):
        words[j] = stemmer.stem(words[j])
        trie_insert(root_, words[j], i)
        
    unq, cnt = np.unique(words, return_counts=True)
    vocab_ += unq.tolist()
    tf[i] = dict(zip(unq, cnt))

vocab_ = np.unique(vocab_)
vocab_ = dict(zip(vocab_, range(len(vocab_))))

N_w = len(list(vocab_))
X_tr = np.zeros((N, N_w))
X_vl = np.zeros((N_val, N_w))

for i in range(N):
    for w in list(tf[i]):
        df = trie_search(root_, w)
        X_tr[i][vocab_[w]] = tf[i][w] * np.log10(N / df)

tf_val = [None for _ in range(N_val)]

for i in range(N_val):
    words = (val_data[i]['title']+" "+val_data[i]['body']).lower().translate(remove_punc).split()
    words = [stemmer.stem(w) for w in words]
    uniq, count = np.unique(words, return_counts=True)
    tf_val[i] = dict(zip(uniq, count))
    
    for word in uniq:
        if word in vocab_:
            df = trie_search(root_, word)
            X_vl[i][vocab_[word]] = tf_val[i][word] * np.log10(N / df) 
```

# K-means

```python
import copy

def k_means(X, y, k, max_iter=100):

    mean_vector = np.mean(X, axis=0)
    var_vector = np.var(X, axis=0)
    centroid = (np.random.randn(k, X.shape[1]) * np.sqrt(var_vector)) + mean_vector
    
    cluster = 0
    for iteration in range(max_iter):
        
        cluster = [[] for _ in range(k)]
        labels = [[] for _ in range(k)]
        
        for i in range(X.shape[0]):
            min_distance = np.inf
            index = 0
            for j in range(k):
                distance = np.linalg.norm(X[i] - centroid[j])
                if distance < min_distance:
                    min_distance = distance
                    index = j
                    
            cluster[index].append(X[i])
            labels[index].append(y[i])
        
        pre_centroid = copy.deepcopy(centroid)
        for i in range(k):
            centroid[i] = np.sum(cluster[i], axis=0) / len(cluster[i])
            
        dif = np.sum(np.linalg.norm(np.array(centroid) - np.array(pre_centroid), axis=1))
        print("Interation num: ", iteration, " sum of Euclidian distance of pre and next centroids:", dif)
        if dif < 1e-3:
            break
    return cluster, labels
```
